```python

def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data1['genres_list'] = data1['genres'].map(lambda x: make_genresList(x))
```

```python

def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data1['actor_1_name'] = data1['cast'].map(lambda x: get_actor1(x))
```

```python

def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data1['actor_2_name'] = data1['cast'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data1['actor_3_name'] = data1['cast'].map(lambda x: get_actor3(x))
```

```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score,accuracy_score
import pickle
```

```python
import nltk
import ssl

try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context
    
nltk.download("stopwords")
```

```python
dataset = pd.read_csv('reviews.txt',sep = '\t', names =['Reviews','Comments'])
```

```python
stopset = set(stopwords.words('english'))
```

```python
vectorizer = TfidfVectorizer(use_idf = True,lowercase = True, strip_accents='ascii',stop_words=stopset)
```

```python
X = vectorizer.fit_transform(dataset.Comments)
y = dataset.Reviews
pickle.dump(vectorizer, open('tranform.pkl', 'wb'))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=40)
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X_train,y_train)
```

```python
accuracy_score(y_test,clf.predict(X_test))*100
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X,y)
```

```python
accuracy_score(y_test,clf.predict(X_test))*100
```
